```python
import pandas as pd
import numpy as np
from numpy import NaN
```

```python
data=pd.read_csv('../test/CVDNX_PSM_4758prot_21sample_20200918.csv',index_col=0,header=0)
```

```python
data[data==0]=np.nan
data
```

```python
df=np.log2(data)
df
```

```python
group1=[0,1,2,3,4,5,6]
group2=[7,8,9,10,11,12,13]
list=group1+group2
columns=df.columns[list]
df=pd.DataFrame(df,columns=columns).T
```

```python
df['index']=df.index
df['index'][group1]='grp1'
df['index'][group2]='grp2'
```

```python
df.index=df['index']
df=df.drop('index',axis=1)
```

```python
dff=df.applymap(lambda x:2**x if type(x)==float else np.nan)
dff
```

```python
m=np.nanmin(dff.min().values)*0.8
m
```

```python
dff.fillna(m*0.8, inplace=True)
```

```python
print(dff)
```

```python
dff=dff.groupby(dff.index).agg(np.mean).T
dff
```

```python
dff['fd']=np.log2(dff['grp1']/dff['grp2'])
```

```python

df.fillna(0.8*m,inplace=True)
df
```

```python
from scipy import stats


df=df.iloc[:,:2]
```

```python
dff['p_value']=df.apply(stats.ttest_ind(df[df.index=='grp1'],df[df.index=='grp2'],equal_var=False), axis=0)
```
